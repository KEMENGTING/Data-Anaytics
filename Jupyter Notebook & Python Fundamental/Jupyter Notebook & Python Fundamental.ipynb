#### 單維度選取, 只能選取列或欄位其中一邊資料, <font color=blue>透過布林值判斷</font>
* <font color=red>__df[df[欄位名稱]條件]__</font><br>
`df[df[‘column name’] 條件]`<br>
`df[(df[‘column name’] 條件)&(df[‘column name’] 條件)]`<br>
`df[(df[‘column name’] 條件)|(df[‘column name’] 條件)]`<br>

HINT：條件可以是(等於) == , (不等於) != , > , < , >= , <=<br>

#### 多維度選取, 可同時選取列、欄位資料, <font color=blue>透過位置判斷</font>
* __df.loc[列索引,欄位名稱]__<br>
Selecting data by row and column name (.loc)<br>
* __df.loc[df[欄位名稱]條件]__<br>
Selecting data by a conditional statement (.loc)<br>
`df.loc[df[‘column name’] 條件]`<br>

* __df.iloc[列索引,欄位索引]__<br>
Selecting data by row numbers (.iloc)

df.loc[ : , : ]會印出所有資料<br>
df.iloc[ : , : ]會印出所有資料<br>

### 明明都有名字欄位，能不能將其他欄位一起合併?

<font color=red>__★ merge用法__</font><br>
default：pd.merge(left,right,how="inner",on=None,left_on=None,right_on=None,left_index=False,right_index=False,sort=True,
         suffixes=("_x", "_y"),copy=True,indicator=False,validate=None,)<br>
         
<font color=red>__★ 合併資料集要先檢查欲合併的資料集欄位型態、欄位名稱是否一致__</font>



* 藍色是merge後最後有資料的結果
* 綠色是缺失值，最後會以NaNs呈現
* 紅色會被排除在結果外(不會顯示於結果)

Join的方式與SQL相同，分成left(以左邊table為基準)、right(以右邊table為基準)、inner(取兩個table共同的項目)、outer(取兩個table所有的項目)

```python
#合併資料集要先檢查欲合併的資料集欄位型態、欄位名稱是否一致
#因df、df2都有名字屬性欄位但欄位名稱不同(df：Name、df2：姓名)，故合併資料時要改為相同的欄位名稱才有共同的key值予以合併，否則會出現error
df2 = df2.rename(columns={'姓名':'Name'})
```

```python
#根據Name欄位以df資料集為主Left join df2，串接相同欄位內容
pd.merge(df, df2, how="left", on=["Name"])
```

### 群組【groupby、aggfunc】
<a href="https://zhuanlan.zhihu.com/p/101284491"> Link to 聚合懶人包</a><br>
Step1. groupby

Step2. aggregation 依據


```python
#僅groupby不會呈現任何結果，output顯示"這是一個groupby的資料類型"
record.groupby(['OP_DT'])
```

```python
list(record.groupby(['OP_DT']))
```

```python
record.groupby(['OP_DT']).size()
```

```python
record.groupby(['OP_DT']).count()
#size()、count()差別在於size包含NA值資料但counts只呈現有值的資料
#不一樣的LEVEL
```

```python
#將欄位調整成同一個level
record.groupby(['OP_DT']).count().reset_index()
```

```python
#根據OP_DT為主要groupby key值計算各OP_DT的PEOPLE列數
record.groupby(['OP_DT'])['PEOPLE'].count()
```

```python
#根據OP_DT為主要groupby key值計算各OP_DT的唯一PEOPLE數(即各日期/使用門號數)
record.groupby(['OP_DT'])['PEOPLE'].nunique()
```

```python
record.groupby(['OP_DT']).agg({'PEOPLE':'nunique'})
```

```python
record.groupby(['OP_DT']).agg({'PEOPLE':'nunique'}).reset_index()
```

### 樞紐表【pivot_table】

```python
#目標：計算各日、各時段PEOPLE數
#運用樞紐表：列index 放日期, 欄columns 放小時, 計算門號唯一值搭配values、aggfunc
record.pivot_table(index=['OP_DT'],columns=['HOUR'],aggfunc={'PEOPLE':'nunique'},values=['PEOPLE'])
```
